#### ¿Cuántos de estos cumplen con las normas de compliance?


Calculamos el total

```python
# Filtrar productos en estado "Planned" o "Production"
filtered_df = df[df["Status"].isin(["Planned", "Production"])]

# Contar cuántos cumplen y cuántos no cumplen
cumplen = filtered_df["IsCompliance"].sum()
no_cumplen = len(filtered_df) - cumplen

# Porcentaje de cumplimiento
porcentaje_cumplen = (cumplen / len(filtered_df)) * 100

# Resultados
print(f"Total de Data Products analizados: {len(filtered_df)}")
print(f"Cumplen con las normas de compliance: {cumplen}")
print(f"No cumplen con las normas de compliance: {no_cumplen}")
print(f"Porcentaje de cumplimiento: {porcentaje_cumplen:.2f}%")
```

Vamos a verlo desglosado por estado planned y production

```python
# Filtrar los estados que nos interesan
estados_interes = ["Planned", "Production"]
filtered_df = df[df["Status"].isin(estados_interes)]

# Agrupar por 'Status' y calcular las métricas
tabla_resultados = filtered_df.groupby("Status").agg(
    Total_Analizado=("IsCompliance", "count"),
    Cumplen=("IsCompliance", "sum")
).reset_index()

# Calcular 'No Cumplen' y 'Porcentaje Cumplimiento (%)'
tabla_resultados["No Cumplen"] = tabla_resultados["Total_Analizado"] - tabla_resultados["Cumplen"]
tabla_resultados["Porcentaje Cumplimiento (%)"] = (
    tabla_resultados["Cumplen"] / tabla_resultados["Total_Analizado"] * 100
).round(2).astype(str) + "%"

# Calcular los totales acumulados
totales = tabla_resultados[["Total_Analizado", "Cumplen", "No Cumplen"]].sum()
porcentaje_total = (totales["Cumplen"] / totales["Total_Analizado"] * 100) if totales["Total_Analizado"] > 0 else 0

# Crear una fila para los totales
fila_totales = pd.DataFrame({
    "Status": ["Total"],
    "Total_Analizado": [totales["Total_Analizado"]],
    "Cumplen": [totales["Cumplen"]],
    "No Cumplen": [totales["No Cumplen"]],
    "Porcentaje Cumplimiento (%)": [f"{porcentaje_total:.2f}%"]
})

# Concatenar los totales al DataFrame de resultados
tabla_resultados = pd.concat([tabla_resultados, fila_totales], ignore_index=True)

# Renombrar columnas para mejor presentación
tabla_resultados.rename(columns={
    "Status": "Estado",
    "Total_Analizado": "Total Analizado",
    "Cumplen": "Cumplen",
    "No Cumplen": "No Cumplen",
    "Porcentaje Cumplimiento (%)": "Porcentaje Cumplimiento (%)"
}, inplace=True)

# Ordenar las columnas si es necesario
tabla_resultados = tabla_resultados[["Estado", "Total Analizado", "Cumplen", "No Cumplen", "Porcentaje Cumplimiento (%)"]]

# Mostrar la tabla
tabla_resultados
```

La tabla muestra un análisis de cumplimiento en dos estados, "Planned" y "Production". En el estado "Planned", el porcentaje de cumplimiento es del 47.06%, mientras que en "Production" es notablemente mayor, con un 64.52%. En total, se han analizado 113 casos, de los cuales 64 cumplen y 49 no cumplen, lo que da un porcentaje de cumplimiento global del 56.64%. Esto sugiere que el estado de producción tiene un desempeño significativamente mejor que el estado planificado en términos de cumplimiento.

```python
# Agrupar por dominio, owner y estado, contando Data Products
owner_status_counts = df.groupby(['Domain', 'Owner', 'Status'])['Name'].count().reset_index(name='Count')

# Obtener el total de Data Products por owner y dominio
owner_totals = owner_status_counts.groupby(['Domain', 'Owner'])['Count'].sum().reset_index(name='TotalCount')

# Unir los DataFrames
merged_df = pd.merge(owner_status_counts, owner_totals, on=['Domain', 'Owner'])

# Filtrar solo los registros donde el estado es "Deprecated"
deprecated_df = merged_df[merged_df['Status'] == 'Deprecated'].copy()

# Calcular el porcentaje de Data Products en estado "Deprecated"
deprecated_df['DeprecatedPercentage'] = ((deprecated_df['Count'] / deprecated_df['TotalCount']) * 100).round(2)

# Encontrar el Owner menos productivo en cada dominio (mayor porcentaje de Deprecated)
max_deprecated = deprecated_df.groupby('Domain')['DeprecatedPercentage'].transform('max')
least_productive_owners = deprecated_df[deprecated_df['DeprecatedPercentage'] == max_deprecated]

# Mostrar resultados
print("Owner menos productivo en cada dominio:")
least_productive_owners = least_productive_owners[['Domain', 'Owner', 'DeprecatedPercentage']]
least_productive_owners
```
